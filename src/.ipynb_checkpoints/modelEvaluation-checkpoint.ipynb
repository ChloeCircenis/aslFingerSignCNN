```python
#imports
import pandas as pd
from PIL import Image
import io
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau, TensorBoard
import datetime, os
from sklearn.metrics import classification_report, confusion_matrix
import string
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import model
model = tf.keras.models.load_model("../models/asl_model_50epoch.h5")
```

```python
#import test dataset
data = np.load("../data/test_dataset.npz")
X_test = data["X_test"]
y_test = data["y_test"]

test_dataset = (
    tf.data.Dataset.from_tensor_slices((X_test, y_test))
    .batch(32)
    .prefetch(tf.data.AUTOTUNE)
)
```

```python
#evalute on test set
test_loss, test_acc = model.evaluate(test_dataset)
print(f"Test loss: {test_loss:.4f}, Test accuracy: {test_acc:.4f}")
```

```python
#confusion matrix setup and model evaluation by label
y_true = []
y_pred = []

for x_batch, y_batch in test_dataset:
    preds = model.predict(x_batch, verbose=0)
    y_true.extend(np.argmax(y_batch.numpy(), axis=1))
    y_pred.extend(np.argmax(preds, axis=1))

labels = list(string.ascii_uppercase) 

y_true_letters = [labels[i] for i in y_true]
y_pred_letters = [labels[i] for i in y_pred]

print(classification_report(y_true_letters, y_pred_letters, digits=4))
```

```python
#pretty confusion matrix 
cm = confusion_matrix(y_true_letters, y_pred_letters, labels=labels)

plt.figure(figsize=(14, 10))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=labels,
    yticklabels=labels,
    linewidths=0.5,
    cbar=True
)

plt.title("ASL Letter Classification Confusion Matrix", fontsize=16)
plt.xlabel("Predicted Label", fontsize=14)
plt.ylabel("True Label", fontsize=14)
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
