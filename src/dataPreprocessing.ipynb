```python
#imports
import pandas as pd
from PIL import Image
import io
import numpy as np
import tensorflow as tf
```

```python
#import data
data = pd.read_parquet("../data/train-00000-of-00001.parquet")
```

```python
def decode_image(img_dict):
    raw_bytes = img_dict["bytes"]
    return Image.open(io.BytesIO(raw_bytes))

IMG_SIZE = 64   #resize images to be square, can change size

def preprocess_image(img_dict):
    img = decode_image(img_dict)
    # Convert to grayscale
    img = img.convert("L")
    # Resize
    img = img.resize((IMG_SIZE, IMG_SIZE))
    # Convert to normalized numpy array
    arr = np.array(img) / 255.0
    # Add channel dimension (H, W, 1)
    return arr[..., np.newaxis]
```

```python
#set number of classes and split into X and y
num_classes = 26

y = tf.keras.utils.to_categorical(data["label"].values, num_classes=num_classes)
X = np.stack([preprocess_image(im) for im in data["image"]])
```

```python
np.save("../data/X.npy", X)
np.save("../data/y.npy", y)
```
